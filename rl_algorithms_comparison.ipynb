## Overview

This project focuses on comparing the performance of various reinforcement learning (RL) algorithms in the Gymnasium MuJoCo Humanoid-v4 environment. The goal is to evaluate and analyze how different RL models handle the complexities of the Humanoid-v4 task.

## Environment and Frameworks

- **Environment:** Gymnasium's MuJoCo Humanoid-v4
- **Frameworks:** Utilizing Stable Baselines 3 for model implementation

## Tested RL Models

The following RL algorithms have been tested in this project:

1. **Soft Actor-Critic (SAC)**
2. **Proximal Policy Optimization (PPO)**
3. **Advantage Actor-Critic (A2C)**
4. **Twin Delayed DDPG (TD3)**

## Experimental Setup

To ensure a fair comparison, multiple runs were conducted for each RL model. This approach accounts for the inherent instability in RL training, along with the complex dynamics of the MuJoCo Humanoid-v4 environment. The decision to conduct several runs aims to capture the variations and potential drops in performance during training.

## Data Collection

Episode length mean data was collected after training runs for each RL model. After the models have been stopped, the data was downloaded from tensorboard. This metric serves as a key indicator of the model's performance over time.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Run 1


##### Overview

Run 1 of the experiment involved training four different reinforcement learning (RL) models in Gymnasium's MuJoCo Humanoid-v4 environment. The training sessions were conducted for a fixed duration of two hours, after which the models were halted, and training data was downloaded from TensorBoard for analysis.

- **Duration:** Two hours of training for each RL model


Importing Data for epiosode length mean

```python
sac_ep_len_mean = pd.read_csv('Run Data/Run 1/SAC/Ep_len_mean/SAC_0.csv')
a2c_ep_len_mean = pd.read_csv('Run Data/Run 1/A2C/ep_len_mean/A2C_0.csv')
ppo_ep_len_mean = pd.read_csv('Run Data/Run 1/PPO/ep_len_mean/PPO_0.csv')
td3_ep_len_mean = pd.read_csv('Run Data/Run 1/TD3/ep_len_mean/TD3_0.csv')
```

<a id='Episode-Length-Mean'></a>
[Back to Top](#Table-of-Contents)

Plotting the episode length mean for run 1

```python
plt.figure(figsize=(10, 6))

plt.plot(sac_ep_len_mean['Step'], sac_ep_len_mean['Value'], label='SAC', linewidth=2)
plt.plot(a2c_ep_len_mean['Step'], a2c_ep_len_mean['Value'], label='A2C', linewidth=2)
plt.plot(ppo_ep_len_mean['Step'], ppo_ep_len_mean['Value'], label='PPO', linewidth=2)
plt.plot(td3_ep_len_mean['Step'], td3_ep_len_mean['Value'], label='TD3', linewidth=2)

# Adding labels and title
plt.xlabel('Steps')
plt.ylabel('Episode Length Mean')
plt.title('Episode Length Mean for Run 1\n(two hours of runtime)')

# Adding legend
plt.legend()

# Display the plot
plt.show()
```

# Analysis of Episode Mean Length Plot

## Interpretation of Results

The plot depicting episode mean length for the four RL models over the two-hour runtime reveals interesting insights into their performance and computational efficiency.

### Step Discrepancy

Observing the different step counts for each model, it becomes apparent that algorithms like TD3 and SAC have significantly fewer steps compared to A2C and PPO. This discrepancy may suggest varying levels of computational complexity among the algorithms. TD3 and SAC could be more computationally expensive, leading to fewer steps within the given two-hour runtime. Conversely, A2C and PPO exhibit a higher step count, hinting at potentially lower computational requirements.

### Model-Specific Considerations

The step count divergence may also indicate inherent differences between the models. Further investigation is required to understand why certain algorithms, such as SAC and TD3, make fewer steps. This observation highlights the need for hyperparameter optimization in subsequent runs to enhance the computational efficiency of SAC and TD3.

### Performance Analysis

Notably, despite the differing step counts, SAC demonstrates superior performance with a considerably lower episode mean length. On the other hand, A2C and TD3 appear to perform less optimally, suggesting potential areas for improvement. This discrepancy underscores the importance of model-specific hyperparameter tuning to unlock the full potential of each algorithm in the challenging MuJoCo Humanoid-v4 environment.

## Future Steps

1. **Hyperparameter Optimization:** Addressing the observed disparities in step counts and performance, the next runs will include thorough hyperparameter optimization for SAC and TD3 to enhance their computational efficiency and overall effectiveness.

2. **Performance Tuning:** Fine-tuning hyperparameters for A2C and TD3 will be a focus to elevate their performance and bridge the gap with SAC.

3. **Further Analysis:** Subsequent runs will delve deeper into the reasons behind the step count differences and performance variations, refining the understanding of each algorithm's behavior in the MuJoCo Humanoid-v4 environment.

# Episode Length Mean per Step

## Purpose of the Second Plot

The decision to create a second plot arises from the inherent variations in step counts across the different RL models during the two-hour training runtime. The initial episode mean length plot, while informative, does not provide a clear comparison on a step-to-step basis due to the disparity in the number of steps taken by each model.

```python
min_max_steps = min(
    sac_ep_len_mean['Step'].max(),
    a2c_ep_len_mean['Step'].max(),
    ppo_ep_len_mean['Step'].max(),
    td3_ep_len_mean['Step'].max()
)

# Create copies of DataFrames with rows up to the minimum number of steps
sac_ep_len_mean_shortened = sac_ep_len_mean[sac_ep_len_mean['Step'] <= min_max_steps].copy()
a2c_ep_len_mean_shortened = a2c_ep_len_mean[a2c_ep_len_mean['Step'] <= min_max_steps].copy()
ppo_ep_len_mean_shortened = ppo_ep_len_mean[ppo_ep_len_mean['Step'] <= min_max_steps].copy()
td3_ep_len_mean_shortened = td3_ep_len_mean[td3_ep_len_mean['Step'] <= min_max_steps].copy()

# Plotting
plt.figure(figsize=(10, 6))

# Plotting shortened DataFrames
plt.plot(sac_ep_len_mean_shortened['Step'], sac_ep_len_mean_shortened['Value'], label='SAC', linewidth=2)
plt.plot(a2c_ep_len_mean_shortened['Step'], a2c_ep_len_mean_shortened['Value'], label='A2C', linewidth=2)
plt.plot(ppo_ep_len_mean_shortened['Step'], ppo_ep_len_mean_shortened['Value'], label='PPO', linewidth=2)
plt.plot(td3_ep_len_mean_shortened['Step'], td3_ep_len_mean_shortened['Value'], label='TD3', linewidth=2)

# Adding labels and title
plt.xlabel('Steps')
plt.ylabel('Episode Length Mean')
plt.title('Episode Length Mean for Run 1')

# Adding legend
plt.legend()

# Display the plot
plt.show()
```

# Interpretation of Second Plot: Episode Length Mean per Step

## Observations

Upon analyzing the second plot detailing episode length mean per step, distinct performance trends emerge among the tested RL models.

### SAC Performance

SAC (Soft Actor-Critic) stands out as the top-performing model, demonstrating a steady and efficient learning trajectory. Despite a slower pace, SAC exhibits notable step-to-step improvement, suggesting a robust adaptation to the complexities of the MuJoCo Humanoid-v4 environment.

### TD3 Challenges

In contrast, TD3 (Twin Delayed DDPG) faces challenges, seemingly unable to achieve significant results. Notably, insights from the implementation's author suggest that historical testing did not include the Humanoid environment. Additionally, a bug related to non-integer max actions, as in the Humanoid environment, may have hindered TD3's performance.

## Author's Note

The author of the TD3 implementation in Stable Baselines acknowledged potential limitations in testing, indicating that the model might not have been optimized for the Humanoid environment at the time of development. The non-integer max action bug further complicates TD3's suitability for the specific action range used by Humanoid.

[Citation: (Author's Note on TD3 Implementation in Stable Baselines)](https://github.com/sfujim/TD3/issues/19)

# A2C Loss

```python
entropy_loss = pd.read_csv('Run Data/Run 1/A2C/entropy_loss/A2C_0.csv')
policy_loss = pd.read_csv('Run Data/Run 1/A2C/policy_loss/A2C_0.csv')
value_loss = pd.read_csv('Run Data/Run 1/A2C/value_loss/A2C_0.csv')

# Plotting Entropy Loss
plt.figure(figsize=(10, 6))
plt.plot(entropy_loss['Step'], entropy_loss['Value'], label='Entropy Loss', color='blue')
plt.xlabel('Steps')
plt.ylabel('Entropy Loss')
plt.title('Entropy Loss for A2C')
plt.legend()
plt.grid(True)
plt.show()

# Plotting Policy Loss
plt.figure(figsize=(10, 6))
plt.plot(policy_loss['Step'], policy_loss['Value'], label='Policy Loss', color='green')
plt.xlabel('Steps')
plt.ylabel('Policy Loss')
plt.title('Policy Loss for A2C')
plt.legend()
plt.grid(True)
plt.show()

# Plotting Value Loss
plt.figure(figsize=(10, 6))
plt.plot(value_loss['Step'], value_loss['Value'], label='Value Loss', color='red')
plt.xlabel('Steps')
plt.ylabel('Value Loss')
plt.title('Value Loss for A2C')
plt.legend()
plt.grid(True)
plt.show()
```

## Interpretation

### Convergence to Optimal Policy:
The decreasing entropy loss, along with the stabilizing policy loss and value loss, suggests that the A2C algorithm is converging towards an optimal or near-optimal policy. The decreasing entropy loss indicates a shift towards deterministic behavior, while the stabilizing policy and value losses imply that the learned policy and value function are becoming increasingly accurate and reliable.

### Balanced Exploration-Exploitation:
The decreasing entropy loss reflects a balance between exploration and exploitation. As the entropy loss decreases, the agent's actions become more exploitative, leveraging its learned knowledge to maximize rewards while maintaining sufficient exploration to avoid premature convergence to suboptimal solutions.

### Training Progress:
The trends observed in the latter half of the training steps indicate that the A2C algorithm is making significant progress towards learning an effective policy and value function. The stabilizing losses suggest that the algorithm has reached a state of relative stability, where further training may yield diminishing returns or marginal improvements in performance.
